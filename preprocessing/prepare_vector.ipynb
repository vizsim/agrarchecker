```python
import subprocess

input_fgb = "CTM_GER_2022_seg_v201_4326.fgb"

cmd = [
    "tippecanoe",
    "-o", "CTM_GER_2022_seg_v201.pmtiles",
    "--minimum-zoom=11",
    "--maximum-zoom=12",
    "--drop-rate=0",
    "--drop-densest-as-needed",
    "--no-feature-limit",
    "--no-tile-size-limit",
    "--maximum-tile-bytes=1000000",
    "--force",
    "-l", "poly",
    input_fgb
]

print("Running tippecanoe ...")
subprocess.run(cmd, check=True)
```
